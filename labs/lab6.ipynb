```python
from pathlib import Path
import importlib.util
import sys

module_candidates = [
    Path.cwd() / 'generate_lab6.py',
    Path.cwd() / 'lab6' / 'generate_lab6.py',
]
module_path = next(path for path in module_candidates if path.exists())

spec = importlib.util.spec_from_file_location('generate_lab6', module_path)
generate_lab6 = importlib.util.module_from_spec(spec)
sys.modules[spec.name] = generate_lab6
spec.loader.exec_module(generate_lab6)
summary = generate_lab6.run_pipeline()
summary
```
